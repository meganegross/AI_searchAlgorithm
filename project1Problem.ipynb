Read in data for the variable containing name and coordinate information, create a tree.

```python
nodes=[]
node=[]
name=coordinateData["locationName"]
location=(coordinateData['x'], coordinateData['y'])

for x1,x2 in zip(coordinateData["locationName"],zip(coordinateData["x"], coordinateData["y"])):
  node.append(x1)
  node.append(x2)
  nodes.append(node)
  node=[]

print ('Node data: {}'.format(nodes))
```

Read in data for the variable containing neighbor relations between locations, create a tree.

```python
totalRelations=[]
relations = []
relation = []

for i in range (neighborRelationsData.shape[0]-1):
    x = neighborRelationsData.iloc[i,:]
    relationKey=(x[0])
    for j in range (1,neighborRelationsData.shape[1]):
        if (j>i and x[j] != 0 ):
            relation.append([neighborRelationsData.columns[j],x[j]])
    if (len(relation) != 0):
        relations = [relationKey, relation]
        totalRelations.append(relations)

    relation =[]
    relations=[]

print('Node neighbors data: {}'.format(totalRelations))
```
